```python
import re
import numpy as np
import math


def unique_file(input_filename, output_filename):
    input_file = open(input_filename, 'r')
    #file_contents = input_file.read()
    
    strr=""
    count=0
    for line in input_file:
        strr=strr+line
        count+=1
    input_file.close()
    word_list = str.split(strr)
    words=set(word_list)

    file = open(output_filename, 'w')
    file.write(str(words) + "\n")
    file.close()
    return [words,count]

def binary_vsm(x):
    j=0
    #print(np.array(x).shape)
    global lista
    input_file_binary = open("input.txt", 'r')
    for line in input_file_binary:
        #print(line)
        for i in range(0,len(lista)):
            struy="\\b"+lista[i]+"\\b"
            #print(lista[i])
            match= re.findall(struy,line)
            struy=""
            if match:
                x[j][i]=1
        j=j+1
    input_file_binary.close()
    
def frequency_vsm(x):
    j=0
    #print(np.array(x).shape)
    global lista
    input_file_binary = open("input.txt", 'r')
    for line in input_file_binary:
        #print(line)
        for i in range(0,len(lista)):
            struy="\\b"+lista[i]+"\\b"
            #print(lista[i])
            match= re.findall(struy,line)
            struy=""
            if match:
                x[j][i]=len(match)
        j=j+1
    input_file_binary.close()
def caped_vsm(x):
    j=0
    #print(np.array(x).shape)
    global lista
    input_file_binary = open("input.txt", 'r')
    for line in input_file_binary:
        #print(line)
        for i in range(0,len(lista)):
            struy="\\b"+lista[i]+"\\b"
            #print(lista[i])
            match= re.findall(struy,line)
            struy=""
            if match:
                if len(match)==1:
                    x[j][i]=1
                elif len(match)>1:
                    x[j][i]=2
        j=j+1
    input_file_binary.close()
def lowerbound_vsm(x):
    j=0
    #print(np.array(x).shape)
    global lista
    input_file_binary = open("input.txt", 'r')
    for line in input_file_binary:
        #print(line)
        for i in range(0,len(lista)):
            struy="\\b"+lista[i]+"\\b"
            #print(lista[i])
            match= re.findall(struy,line)
            struy=""
            if match:
                if len(match)>4:
                    x[j][i]=len(match)
        j=j+1
    input_file_binary.close()
    
def log_vsm(x):
    j=0
    #print(np.array(x).shape)
    global lista
    input_file_binary = open("input.txt", 'r')
    for line in input_file_binary:
        #print(line)
        for i in range(0,len(lista)):
            struy="\\b"+lista[i]+"\\b"
            #print(lista[i])
            match= re.findall(struy,line)
            struy=""
            if match:
                x[j][i]=(math.log(len(match),10)+1)
        j=j+1
    input_file_binary.close()

a=unique_file("input.txt","output.txt")

lista=list(a[0])
counta=a[1]

alpha=len(lista)
#print(lista)
#print("\n")
x=np.zeros(shape=(counta,alpha), dtype=int)
#print(x.shape)
#vfunc = np.vectorize(binary_vsm)
file="input.txt"
y=x.tolist()
#print(np.array(y).shape)
a=input('''
ENTER 1 FOR FREQUENCY VSM
ENTER 2 FOR BINARY VSM
ENTER 3 FOR LOG VSM
ENTER 4 FOR CAPED
ENTER 5 FOR LOWER BOUND VSM 
''')
if int(a)==1:
    frequency_vsm(y)
elif int(a)==2:
    binary_vsm(y)
elif int(a)==3:
    log_vsm(y)
elif int(a)==4:
    caped_vsm(y)
elif int(a)==5:
    lowerbound_vsm(y)

print(y)
```
