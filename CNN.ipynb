```python
from keras.models import Sequential
from keras.layers.core import Dropout,Dense,activations,Flatten
from keras.layers.convolutional import MaxPooling2D,Convolution2D
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
import keras
```

```python
df=pd.read_csv('/home/talha/Documents/mnist.csv')
```

```python
df.iloc[3,1:].values.reshape(28,28).astype('uint8')
```

```python
df_x=df.iloc[:,1:].values.reshape(len(df),28,28,1)
df_y=df.iloc[:,0].values
df_y
```

```python
df_y=keras.utils.to_categorical(df_y,num_classes=10)
df_y
```

```python
df_x=np.array(df_x)
df_y=np.array(df_y)
df_y
```

```python
df_x.shape
```

```python
df_y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_x,df_y,test_size=0.2,random_state=4)
```

```python
model=Sequential()
model.add(Convolution2D(32,3,data_format='channels_last',activation='relu',input_shape=(28,28,1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(100))
model.add(Dropout(0.5))
model.add(Dense(10))
#model.add(Activation('softmax'))

model.compile(loss='categorical_crossentropy',optimizer='adadelta',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test))
```

```python
model.predict(x_test[1:20])
```

```python
model.evaluate(x_test,y_test)
```
